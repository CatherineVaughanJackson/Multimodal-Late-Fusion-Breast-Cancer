# Finding Second Contrast Folders and Moving Dicom to New Folder

**file directory layout**

Manifest -> Duke_breast_cancer -> patient_ID -> inter_folder -> different_contrast_folders -> dicom files

```python
main_dir = r"D:\Catherine\Duke_Images\manifest-1657234763605\Duke-Breast-Cancer-MRI"
new_dir = r"D:\Catherine\Test_run"
```

```python
for patient in Patient_Id_list:

    #new directory
    #make new folder for files if not already done so 
    new_dir_ID = os.path.join(new_dir, patient)
    
    if not os.path.exists(new_dir_ID):
        os.makedirs(new_dir_ID)    

        #old directory
        #add patient_id dir
        patient_dir = os.path.join(main_dir, patient)

        #add unique interfolder name
        inter_folder = os.listdir(patient_dir)
        inter_folder_dir = os.path.join(patient_dir, inter_folder[0])
        #print(inter_folder_dir)

        #search through contrast_folders for Ph2ax or 2nd pass
        contrast_scan_folders = os.listdir(inter_folder_dir)
        #print(contrast_scan_folders)
        
        final_folder = None
        
        for folder in contrast_scan_folders:
            if "Ph2ax" in folder:
                final_folder = folder
            elif "2nd pass" in folder:
                final_folder = folder
            elif "Ph2Ax" in folder:
                final_folder = folder
            else:
                pass
        
        if final_folder is None:
            print(patient, "folder not found")
        else:
            
            #final_directory to dicom files 2nd contrast
            final_directory = os.path.join(inter_folder_dir, final_folder)
            print(final_directory)


            #dicom file directories 
            dicom_files = os.listdir(final_directory)

            #copy each file into new folder 
            for file in dicom_files:
                source = os.path.join(final_directory, file)
                destination= os.path.join(new_dir_ID, file)
                try:
                    shutil.copy(source, destination)
                except:
                    print("error occured", file)
            
            print(patient, "successfully copied")
            old_folder_length = len(os.listdir(final_directory))
            new_folder_length = len(os.listdir(new_dir_ID))
            if old_folder_length != new_folder_length:
                print("files have been lost", old_folder_length, ":", new_folder_length)
            else:
                print(new_folder_length, "slices")
    else:
        print(patient, "files already coppied")
```
